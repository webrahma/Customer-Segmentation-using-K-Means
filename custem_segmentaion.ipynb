```python
features=['Age','Income','Total_spending','NumWebPurchases','NumStorePurchases','NumWebVisitsMonth','Recency']
```

```python
x=data[['Age','Income','Total_spending','NumWebPurchases','NumStorePurchases','NumWebVisitsMonth','Recency']].copy()
```

```python
x
```

```python
from sklearn.preprocessing  import  StandardScaler
scaler=StandardScaler()
```

```python
x_scale=scaler.fit_transform(x)
```

```python
x_scale
```

```python
from sklearn.cluster import KMeans
```

```python
#we will create elpow method to find the best N-of-k
wcss=[]
for i in range (2,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x_scale)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10), wcss, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans=KMeans(n_clusters=6)
data['Cluster'] =kmeans.fit_predict(x_scale)
```

```python
data.head()
```

```python
cluster_summery=data.groupby('Cluster')[features].mean()
```

```python
cluster_summery
```

```python
data['Cluster'].value_counts()
```

```python
#to drow plot to cluster must minmize numper of feature so will user pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_data=pca.fit_transform(x_scale)
data['pca1'],data['pca2']=pca_data[:,0],pca_data[:,1]
```

```python
sns.scatterplot(x='pca1', y='pca2', hue='Cluster', data=data, palette='Set1')
plt.title("Customer Segments after PCA")
plt.show()
```
